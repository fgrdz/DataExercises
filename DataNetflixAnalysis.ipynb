```python
#inicio do projeto
import pandas as pd
import numpy as np

arquivo = '/content/drive/MyDrive/titles.csv'

dataframe = pd.read_csv(arquivo)
dataframe = dataframe[['title', 'type', 'release_year', 'genres']]
dataframe
```

```python
from collections import Counter

genres_movie = dataframe[dataframe['type']=='MOVIE']['genres'].str.split(',')

genre_counts = genres_movie.explode().value_counts()
most_genre = Counter.most_common(genre_counts)
genre_counts

```

```python
#abaixo Ã© possÃ­vel ver que os generos de drama sÃ£o os mais recorrentes
print(list(most_genre))
```
