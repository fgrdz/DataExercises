```python
#inicio do projeto
import pandas as pd
import numpy as np

arquivo = '/content/drive/MyDrive/titles.csv'

dataframe = pd.read_csv(arquivo)
dataframe = dataframe[['title', 'type', 'release_year', 'genres', 'imdb_score']]
dataframe
```

```python
from collections import Counter

genres_movie = dataframe[dataframe['type']=='MOVIE']['genres'].str.split(',')

genre_counts = genres_movie.explode().value_counts()
most_genre = Counter.most_common(genre_counts)
genre_counts

```

```python
#abaixo é possível ver que os generos de drama são os mais recorrentes
print(list(most_genre))
```

```python
#função para extrair as décadas
def decadas(ano):
  return str(ano)[:3]+'0s'

#armazenar as décadas
decades = dataframe['release_year'].apply(decadas)
data_frequency = decades.value_counts()

print(f'Aqui está a quantidade de itens por década: \n{data_frequency}')
```

```python
# verificar se possui mais filmes ou shows

types = dataframe['type'].value_counts()

fig = px.pie(types, names= types.index, values= types,
             title= 'Distribuição entre filmes e séries no catálogo netflix')
fig.update_layout(legend=dict(title='Categorias', orientation="v", x=0.9, y=0.5))
fig.show()
```
